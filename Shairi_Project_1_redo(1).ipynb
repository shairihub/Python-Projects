I want to change out some objects to categories, floats, and intergers for easier data handling. In my original file, not changing these caused errors and issues. I did some research and updated the data types but also used some help from AI to fix errors that I could not figure out and also had it explain to me what the issue with the code was.

```python
df = pd.read_csv('https://raw.githubusercontent.com/CunyLaguardiaDataAnalytics/datasets/master/2014-15_To_2016-17_School-_Level_NYC_Regents_Report_For_All_Variables.csv')

categorical_cols = ['School Name', 'School Level', 'Regents Exam', 'Year']

for c in categorical_cols:
    df[c] = df[c].astype('category')

df['Percent Scoring 80 or Above'] = pd.to_numeric(df['Percent Scoring 80 or Above'], errors='coerce')

df['Percent Scoring Below 65'] = pd.to_numeric(df['Percent Scoring Below 65'], errors='coerce')

drop_columns = ['School DBN', 'Mean Score', 'Number Scoring 65 or Above',
                'Number Scoring Below 65', 'Number Scoring 80 or Above',
                'Percent Scoring 65 or Above', 'Number Scoring CR',
                'Percent Scoring CR']

df = df.drop(columns=drop_columns)

print(df.dtypes)
```

I want to display graphs that show the amount of student scoring below 65 (fail) vs the amount of students scoring above 80 (highest grades). How many students are doing outstanding vs how many are not passing and need improvement in certian areas?

```python
ffig, axes = plt.subplots(1, 2, figsize=(10, 5))

sns.histplot(df['Percent Scoring Below 65'], bins=11, kde=True, ax=axes[0])
axes[0].set_title('Students Scoring Below 65')

sns.histplot(df['Percent Scoring 80 or Above'], bins=11, kde=True, ax=axes[1])
axes[1].set_title('Students Scoring 80 or Above')

plt.tight_layout()
plt.show()
```

I originally was going to compare data from 2015 and 2016 but decided to keep it simple and just do 2015.

```python
schools_to_plot = ['ACORN Community High School', 'Academy for Conservation and the Environment']
years_to_plot = [2015]

filtered_data_for_plot = df[
    (df['Regents Exam'] == 'Common Core English') &
    (df['School Name'].isin(schools_to_plot)) &
    (df['Year'].isin(years_to_plot))
]

ffig, axes = plt.subplots(1, 2, figsize=(12, 6))

sns.histplot(filtered_data_for_plot['Percent Scoring Below 65'], bins=11, kde=True, ax=axes[0])
axes[0].set_title('Common Core English: Percent Scoring Below 65')
axes[0].set_xlabel('Percent Below 65')
axes[0].set_ylabel('Frequency')

sns.histplot(filtered_data_for_plot['Percent Scoring 80 or Above'], bins=11, kde=True, ax=axes[1])
axes[1].set_title('Common Core English: Percent Scoring 80 or Above')
axes[1].set_xlabel('Percent 80 or Above')
axes[1].set_ylabel('Frequency')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

When we examine the english score list closer we can see that the most frequent results were around 20 percent of students scored below 65 in english and 25 percent of students scored above 80.
